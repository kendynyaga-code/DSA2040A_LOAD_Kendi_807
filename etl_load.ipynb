```python
import pandas as pd
import os
import sqlite3 

# Load transformed data
df_full = pd.read_csv('transformed/transformed_full.csv')
df_inc = pd.read_csv('transformed/transformed_incremental.csv')

# Ensure the 'loaded' directory exists
os.makedirs('loaded', exist_ok=True)

#Connect to SQLite database 
conn = sqlite3.connect('loaded/full_data.db') 

#load data into SQLite database
df_full.to_sql('full_data', conn, if_exists='replace', index=False)
df_inc.to_sql('incremental_data', conn, if_exists='replace', index=False)

# Verify the load
pd.read_sql('SELECT * FROM full_data LIMIT 5', conn)

#Preview output

# Display 5â€“10 sample rows
preview_df = pd.read_sql('SELECT * FROM full_data LIMIT 10', conn)
print("Preview of loaded data:")
display(preview_df)

# Confirm column names
print("\nColumn Names:")
print(preview_df.columns.tolist())

# Confirm record count
count_df = pd.read_sql('SELECT COUNT(*) AS total_records FROM full_data', conn)
print("\nTotal Records in Table:")
display(count_df)

# Close the database connection
conn.close()
```
